```python
all_data = [] # collect row here

for i in range(1, 510):
    try: 
        response = requests.get(f"https://api.themoviedb.org/3/movie/top_rated?api_key=073cfe0d7c3d204ff6814a05f45940a7&language=en-US&page={i}")
        response.raise_for_status() # raises error if request failed
        
        data = response.json().get('results', [])
        if not data:
            break  # Stop if there are no more results

        # Extract required columns
        temp_df = pd.DataFrame(data)[['id', 'title', 'overview', 'release_date', 'popularity', 'vote_average', 'vote_count']]
        all_data.append(temp_df)

        time.sleep(0.2)  # Be polite to the API (adjust if rate-limited)

    except Exception as e:
        print(f"Error on page {i}: {e}")
        break


# combine all data
df = pd.concat(all_data, ignore_index = True)

    
```
