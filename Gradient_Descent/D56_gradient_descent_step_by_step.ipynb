```python
from sklearn.datasets import make_regression
import numpy as np
```

```python
X, y = make_regression(n_samples = 4, n_features = 1, n_informative=1, n_targets=1, noise=80, random_state=13)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X, y)
```

```python
# Lets apply OLS
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
reg.fit(X, y) 
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
plt.scatter(X, y)
plt.plot(X, reg.predict(X), color = 'red')
```

```python
# Lets apply Gradient Descent assuming slope is constant m = 78.35
# and lets assume the starting value of intercept b = 0
y_pred = ((78.35 * X) + 0).reshape(4)
```

```python
plt.scatter(X, y)
plt.plot(X, reg.predict(X), color = 'red', label = 'OLS')
plt.plot(X, y_pred, color = 'green', label = 'b = 0')
plt.legend()
plt.show()
```

```python
# Iteration I
m = 78.35
b = 0

loss_slope = -2 * np.sum(y - m * X.ravel() - b)
loss_slope
```

```python
# Lets take learning rate 0.1
lr = 0.1

step_size = loss_slope * 0.1
step_size
```

```python
# calculating the new intercept
b = b - step_size
b
```

```python
y_pred1 = ((78.35 * X) + b).reshape(4)
```

```python
plt.scatter(X, y)
plt.plot(X, reg.predict(X), color = 'red', label = 'OLS')
plt.plot(X, y_pred1, color = 'green', label = "b = {}".format(b))
plt.plot(X, y_pred, color = '#A3E4D7', label = 'b = 0')
plt.legend()
plt.show()
```

```python
# Iteration II
loss_slope1 = -2 * np.sum(y - m * X.ravel() - b)
loss_slope1
```

```python
# Lets take learning rate 0.1
lr = 0.1

step_size2 = loss_slope1 * lr
step_size2
```

```python
# Calculating the new intercept
b1 = b - step_size2
b1
```

```python
y_pred2 = ((78.35 * X) + b1).reshape(4)
```

```python
plt.scatter(X, y)
plt.plot(X, reg.predict(X), color = 'red', label = 'OLS')
plt.plot(X, y_pred2, color = 'green', label = 'b = {}'.format(b1))
plt.plot(X, y_pred1, color = 'blue', label = "b = {}".format(b))
plt.plot(X, y_pred, color = '#A3E4D7', label = 'b = 0')
plt.legend()
plt.show()
```

```python
# Iteration III
loss_slope2 = -2 * np.sum(y - m * X.ravel() - b1)
loss_slope2
```

```python
# lets take learning rate 0.1
lr = 0.1
step_size3 = loss_slope2 * lr
step_size3
```

```python
# calculating the new intercept
b2 = b1 - step_size3
b2
```

```python
y_pred3 = ((78.35 * X) + b2).reshape(4)
```

```python
plt.scatter(X, y)
plt.plot(X, reg.predict(X), color = 'red', label = 'OLS')
plt.plot(X, y_pred3, color = 'yellow', label = 'b = {}'.format(b2))
plt.plot(X, y_pred2, color = 'green', label = 'b = {}'.format(b1))
plt.plot(X, y_pred1, color = 'blue', label = "b = {}".format(b))
plt.plot(X, y_pred, color = '#A3E4D7', label = 'b = 0')
plt.legend()
plt.show()
```

```python
# Perform these in loops 
b = 100
m = 78.35
lr = 0.1

epochs = 10

for i in range(epochs):
    loss_slope = -2 * np.sum(y - m * X.ravel() - b)

    b = b - (lr * loss_slope)

    y_pred = m * X + b

    plt.plot(X, y_pred)

plt.scatter(X, y)
```
