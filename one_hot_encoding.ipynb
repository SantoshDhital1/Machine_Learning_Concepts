# 4. OneHotEncoding using Top Categories

```python
counts = df['brand'].value_counts()
```

```python
df['brand'].nunique()
threshold = 100
```

```python
replaces = counts[counts <= threshold].index
```

```python
replaces
```

```python
pd.get_dummies(df['brand'].replace(replaces, 'uncommon') )
```
